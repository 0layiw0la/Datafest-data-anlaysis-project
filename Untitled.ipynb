Data transformation

```python
# Converting the time stamp column in the crop data frame to the appropriate datatype------------------------------------
crop_dataframe['TIMESTAMP'] = pd.to_datetime(crop_dataframe['TIMESTAMP'])
# ------------------------------------------------------
# Create a new column 'month' that stores the month names-----------------------------------------------------
crop_dataframe['month'] = crop_dataframe['TIMESTAMP'].dt.strftime('%B')
# -----------------------------------------------------------
# Converting the time stamp column in the sensor data frame to the appropriate datatype------------------------------------
sensor_dataframe['TIMESTAMP'] = pd.to_datetime(sensor_dataframe['TIMESTAMP'])
# -----------------------------------------------------------------------------------------
pest_dataframe['TIMESTAMP'] = pd.to_datetime(pest_dataframe['TIMESTAMP'])
pest_dataframe['month'] = pest_dataframe['TIMESTAMP'].dt.strftime('%B')
```

```python
# Creating weights for pest severity------------------------------------------------------------------------------------------
severity_weights = {'Low': 1, 'Medium': 2, 'High': 3, 'NA':None}
pest_dataframe['SEVERITY_WEIGHT'] = pest_dataframe['PEST_SEVERITY'].map(severity_weights) 
# This code maps through pest severity column assigning the preset weights to the severity weight column
# this should allow us see the impact of each pest type more clearly
```

<h1><center>CROP YIELD ANALYSIS</center></h1>

<h2><center><strong><u>AVERAGE YIELD PER CROP</u></strong></center></h2>

- On average there is not much difference in crop yield between the crops, with each one having an average between 549 and 550 bushels per acre. An average crop yield of about 550 bu/ac would be considered exceptionally high for both the low yield(carrot) and high yield(sugarcane,corn,cotton,barely,wheat) crops.



- This high yield could be as a result of exceptional farming practices, favorable weather conditions, and possibly the use of advanced crop management techniques or specialized crop varieties. It could also be due to data inaccuracies, there may be need to verify the accuracy of this data.

```python
# Looking at the mean crop yield of each crop.
crop_dataframe[['CROP_TYPE','CROP_YIELD']].groupby('CROP_TYPE').mean().sort_values('CROP_YIELD',ascending=False).plot(kind='bar',logy=True)
```

<h2><center><strong><u>TOTAL YIELD PER CROP</u></strong></center></h2>

- We look at the total yield between September 2022 and 2023(past one year).

- Rice had the highest total yield during this period, producing approximately 115,929,520.95 bushels per acre (bu/ac).

- Wheat had the lowest total yield during this period, producing approximately 114,910,331.85 bushels per acre (bu/ac).

```python
plt.figure(figsize=(12,8))
total_yield = crop_dataframe.groupby('CROP_TYPE')['CROP_YIELD'].sum().sort_values(ascending=False)
total_yield.plot(kind='bar',logy=True,color=['#12cd19']+['#92751F'] * (len(total_yield) - 1))
for i,v in enumerate(total_yield):
    plt.text(i,v,str(v),va='bottom',ha='center',fontweight='bold',fontsize=12)
plt.ylabel('Total crop yield in bu/ac',fontsize=12)
plt.title('TOTAL CROP YIELD PER CROP',fontsize=30,fontweight='bold')
plt.xticks(rotation=0)
plt.show()
```

<h2><center><strong>YIELD VARIANCE</strong></center></h2>

- The month of December had the highest crop yield with 98,432,622.31 bushels per acre of crop being harvested.
- The least amount of crops were harvested in February (88,061,954.46 bushels per acre of crop was harvested).
- On average 96105610.042 bushels per acre of crop was harvested each month.
- Barely,tomatoes and wheat had the highest yield in December.
- Carrots had their highest yield in march.
- Corn and cotton had their highest yield in January.
- Potatoes and soybeans had their highest yield in May.
- Rice had it's highest yield in October.
- Sugarcanes had their highest yield in August.


```python
# General monthly yield variance for all crops
yield_variance_by_time = crop_dataframe.groupby('month')['CROP_YIELD'].sum().sort_values(ascending=False)
display(yield_variance_by_time)

#How much was harvested on average each month
yield_variance_by_time.mean()
```

```python
# Vizualizing monthly yield variance per crop
yield_variance_by_time_per_crop = crop_dataframe.groupby(['CROP_TYPE', 'month'])['CROP_YIELD'].sum().sort_values(ascending=False).reset_index()

# Create a 3x4 grid of subplots
fig, axes = plt.subplots(3,4, figsize=(60, 18))

# Get unique crop types
unique_crop_types = yield_variance_by_time_per_crop['CROP_TYPE'].unique()

# Iterate through the subplots and plot data for each crop type
for i, ax in enumerate(axes.flatten()):
    if i < len(unique_crop_types):
        crop_type = unique_crop_types[i]
        data = yield_variance_by_time_per_crop[yield_variance_by_time_per_crop['CROP_TYPE'] == crop_type]
        ax.plot(data['month'], data['CROP_YIELD'],linewidth=3,color='#92751F')
        ax.set_title(f'Monthly total yield {crop_type}',fontsize=32,fontweight='bold')
        ax.set_xlabel('Month')
        
        ax.set_ylabel('Crop Yield')
         # Set custom fonts for x and y-axis labels
        ax.tick_params(axis='x', labelsize=22)
        ax.tick_params(axis='y', labelsize=22)
        
        # Add data labels to the line plot
        for x, y in zip(data['month'], data['CROP_YIELD']):
            ax.text(x, y, f'{y:.2f}', ha='center', va='bottom', fontsize=22)

# Adjust subplot layout
plt.tight_layout()

# Show the subplots
plt.show()
```

<h2><center><strong>Growth Stage metrics</strong></center></h2>

- The crops spend the largest amount of time in the fruiting,seedling and harvesting stage in that order and the least in the vegitative state.


```python
growth_stage_frequency = crop_dataframe.groupby('CROP_TYPE')['GROWTH_STAGE'].value_counts().reset_index(name='FREQUENCY')
plt.figure(figsize=(16, 6))
plt.bar(
    x=range(len(growth_stage_frequency)),
    height=growth_stage_frequency['FREQUENCY'],
    tick_label=growth_stage_frequency['CROP_TYPE'] + ' - ' + growth_stage_frequency['GROWTH_STAGE'],color='#12cd19'
)
plt.xlabel('Crop Type - Growth Stage')
plt.ylabel('Frequency')
plt.title('Frequency of Growth Stages by Crop Type',fontsize=30,fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

<h1><center>PEST ISSUE ANALYSIS</center></h1>
<h2>PEST FREQUENCY</h2>

- Caterpillars are the most prominent pest type.
- Mites are the least prominent pest type
- Potatoes have the least amount of pest issues, they are affected the most by fungus and least by mites.
- Barely has the highest amount of pest issues, it is affected the most by Aphids and the least by whiteflies.
- Carrots are affected the most by fungus and the least by caterpillars.
- Corn is affected the most by fungus and the least by whiteflies.
- Cotton is affected the most by fungus and the least by Whiteflies.
- Rice is affected the most by fungus and the least by caterpillars.
- Soybeans are affected the most by fungus and the least by aphids.
- Sugarcane is affected the most by caterpillars and the least by whiteflies.
- Tomato is affected the most by caterpillars and the least by fungus.
- Wheat is affected the most by caterpillars and the least by fungus.

```python
pest_frequency = crop_dataframe.groupby(['CROP_TYPE', 'PEST_ISSUE']).size().unstack(fill_value=0)

ax = pest_frequency.plot(kind='bar',logy=True,width=0.8,figsize=(25,8))
plt.title('PEST ISSUE FREQUENCY PER CROP',fontsize=30,fontweight='bold')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom',fontsize=12,fontweight='bold')
plt.ylabel('Frequency of pest issue(logarithmic scale)',fontsize=14,fontweight='bold')
plt.xticks(rotation=0,fontsize=14,fontweight='bold')
# Show the plot
plt.show()
```
